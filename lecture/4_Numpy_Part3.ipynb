## Coins Example

```python
x = np.array([0, 1, 1, 1, 2, 2,2 , 3])
x
```

```python
sns.countplot(x=x)
```

```python
n_bins = 4

plt.hist(x, n_bins, density=True, histtype='bar')

plt.show()
```

```python
mu, sigma = 0, 0.1

X = np.random.normal(mu, sigma, 1000)

plt.hist(X, 30, density=True)

plt.show()
```

```python
mu, sigma = 0, 0.1

X = np.random.normal(mu, sigma, 1000)

count, bins, _ = plt.hist(X, 30, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(bins - mu)**2 / (2*sigma**2)))

plt.show()
```

```python
mean = (0, 3)

cov = [[1, 0], [0, 1]]
```

```python
pts = np.random.multivariate_normal(mean, cov, size=800)
```

```python
plt.plot(pts[:, 0], pts[:, 1], '.', alpha=0.5)
plt.grid()
plt.show()
```

```python
mean = (0, 3)

cov = [[1, 0], [0, 6]]
```

```python
pts = np.random.multivariate_normal(mean, cov, size=800)
```

```python
plt.plot(pts[:, 0], pts[:, 1], '.', alpha=0.5)
plt.axis('equal')
plt.grid()
plt.show()
```

## Is the iris dataset Normal?

```python
def answer_one(): 
    import numpy as np
    import pandas as pd
    from sklearn.datasets import load_iris 
    Iris = load_iris()     
    data = np.c_[Iris.data, Iris.target]
    columns = np.append(Iris.feature_names, ["target"])
    return pd.DataFrame(data, columns=columns)

df = answer_one()
```

```python
x1 = df.iloc[:, 1]

x2 = df.iloc[:, 2]

x3 = df.iloc[:, 3]
```

```python
sns.distplot(x1, hist=True)
```

```python
sns.distplot(x2, hist=True)
```

```python
sns.distplot(x3, hist=True)
```
